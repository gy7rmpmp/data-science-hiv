```python
df_s = df.groupby(['date'])['N'].sum().reset_index()
df_s.columns = ['date','N']
df_s
```

```python
df_s["date"]
```

```python
df_s["date"] = df_s["date"].dt.strftime('%Y.%m')
df_s.info()
```

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np

df_s["redate"] = df_s["date"].astype(float)

lit_spilt_date=[]

for i in range(60): # str(remonth)
    spilt_date = df_s['date'][i].split('.')
    lit_spilt_date += list(np.float_(spilt_date))
    
    if lit_spilt_date[i*2+1] != None:  #ex: [2017.0, 1.0] => 1.0為月份，需調整
        remonth = round(lit_spilt_date[i*2+1]/13,3)
        lit_spilt_date[i*2+1] = remonth
        
    df_s['redate'][i] = lit_spilt_date[i*2]+remonth
    
print(df_s['redate'])

# use historical data to forecast prices after the year 2021
train = df_s[df_s.redate < 2021]
test = df_s[df_s.redate >= 2021]

```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
# predict prices based on date
X_train = train.date[:, np.newaxis]
# we use a log-transform to get a simpler relationship of data to target
y_train = np.log(train.N)

tree = DecisionTreeRegressor(max_depth=5).fit(X_train, y_train)
linear_reg = LinearRegression().fit(X_train, y_train)

# predict on all data
X_all = df_s.date[:, np.newaxis]

pred_tree = tree.predict(X_all)
pred_lr = linear_reg.predict(X_all)
# undo log-transform
N_tree = np.exp(pred_tree)
N_lr = np.exp(pred_lr)
```

```python
plt.figure(figsize=(16,9))
plt.semilogy(train.date, train.N, label="Training data")
plt.semilogy(test.date, test.N, label="Test data")
plt.semilogy(df_s.date, N_tree, label="Tree prediction")
plt.semilogy(df_s.date, N_lr, label="Linear prediction")
plt.xticks(rotation=50 )

plt.legend(loc=3, prop={'size': 13})
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
# predict prices based on date
X_train = train.date[:, np.newaxis]
# we use a log-transform to get a simpler relationship of data to target
y_train = np.log(train.N)

reg = KNeighborsRegressor(n_neighbors=5)
reg.fit(X_train, y_train)
linear_reg = LinearRegression().fit(X_train, y_train)

# predict on all data
X_all = df_s.date[:, np.newaxis]

pred_reg = reg.predict(X_all)
pred_lr = linear_reg.predict(X_all)

# undo log-transform
N_reg= np.exp(pred_reg)
N_lr = np.exp(pred_lr)
```

```python
plt.figure(figsize=(16,9))
plt.semilogy(train.date, train.N, label="Training data")
plt.semilogy(test.date, test.N, label="Test data")
plt.semilogy(df_s.date, N_reg, label="knn prediction")
plt.semilogy(df_s.date, N_lr, label="Linear prediction")
plt.xticks(rotation=50 )

plt.legend(loc=3, prop={'size': 13})
```
